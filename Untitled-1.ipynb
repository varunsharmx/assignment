3. Compare and contrast mutable and immutable objects in python with examples. 

    In Python, objects can be classified as either mutable or immutable based on whether their state can be changed after they are created. Understanding the differences between these two categories is essential for effective programming.

    Mutable Objects-     
    Mutable objects can be changed or modified after their creation. This means that you can alter their content without creating a new object.  
    Examples:    
        1. Lists: You can add, remove, or change elements in a list.   
        2. Dictionaries: You can change values associated with keys or add new key-value pairs.   
        3. Sets: You can add or remove elements from a set.  
        
    Immutable Objects-  
    Immutable objects cannot be changed once they are created. Any modification will result in the creation of a new object. 
    Examples:  
        1. Tuples: Once a tuple is created, you cannot change its elements.  
        2. Strings: Strings cannot be modified; any operation that alters a string will create a new string.  
        3. Frozensets: Like sets, but immutable.  
    
    Summary-  
    In summary, mutable objects can be modified after creation, while immutable objects cannot. This distinction affects how you manage data and memory in Python, influencing performance, safety, and coding style. Understanding these properties helps in choosing the right data structures for your specific needs.

5. Explain the concept of type casting in python with examples. 

    Type casting in Python is the process of converting one data type into another. This is useful when you need to perform operations that require a specific type or when you want to ensure that values are treated consistently.

    Types of Type Casting  
    1. Implicit Type Casting (Coercion):  
    - This occurs automatically when Python converts one data type to another without explicit instruction. It typically happens when a smaller data type is used in an operation with a larger data type.  
    - Example:  
        a = 5         # Integer  
        b = 2.0       # Float  
        result = a + b  # Implicitly converts 'a' to float  
        print(result)  # Output: 7.0 (float)  
    
    2. Explicit Type Casting:  
    - This requires using built-in functions to convert data types. It is done manually and provides more control over the conversion process.  
    - Common functions used for explicit type casting include:
        * int(): Converts a value to an integer.  
        * float(): Converts a value to a float.  
        * str(): Converts a value to a string.  
        * list(): Converts a value to a list.  
        * tuple(): Converts a value to a tuple.  

    Examples of Explicit Type Casting  
    1. Converting to Integer:  
        num_str = "10"  
        num_int = int(num_str)  # Converts string to integer  
        print(num_int)           # Output: 10  

    2. Converting to Float:  
        num_int = 10  
        num_float = float(num_int)  # Converts integer to float  
        print(num_float)             # Output: 10.0  

    3. Converting to String:
        num = 123  
        num_str = str(num)  # Converts integer to string  
        print(num_str)      # Output: "123"  

    4. Converting to List:  
        my_tuple = (1, 2, 3)  
        my_list = list(my_tuple)  # Converts tuple to list  
        print(my_list)             # Output: [1, 2, 3]  

    5. Converting to Tuple:  
        my_list = [1, 2, 3]  
        my_tuple = tuple(my_list)  # Converts list to tuple  
        print(my_tuple)            # Output: (1, 2, 3)  

    Special Cases  

    - Converting Floats to Integers: When converting a float to an integer, the decimal part is truncated (not rounded).  
        float_num = 10.7  
        int_num = int(float_num)  # Converts to integer  
        print(int_num)             # Output: 10  

    - Handling Invalid Conversions: If you try to convert a value that cannot be cast to the desired type, it will raise an error.  
        invalid_str = "hello"  
        try:  
            num = int(invalid_str)  # This will raise a ValueError  
        except ValueError:  
            print("Cannot convert to integer")  # Output: Cannot convert to integer  

6. How to conditional statements work in python? illustrate with examples. 

    Conditional statements in Python allow you to execute certain blocks of code based on specific conditions. The most common conditional statements are if, elif, and else. These statements help control the flow of a program based on whether a condition is true or false.  

    Basic Structure  
    The basic syntax for conditional statements is:  

    if condition:  
        # Code to execute if condition is true  
    elif another_condition:  
        # Code to execute if the first condition is false and this condition is true  
    else:  
        # Code to execute if all previous conditions are false  

    Examples  
    1. Simple If Statement:  
    age = 18  
    if age >= 18:  
        print("You are an adult.")  # This will be printed  

    2. If-Else Statement:  
    temperature = 30  
    if temperature > 25:  
        print("It's a hot day.")  # This will be printed  
    else:  
        print("It's a cool day.")  

    3. If-Elif-Else Statement:  
    score = 85  
    if score >= 90:  
        print("Grade: A")  
    elif score >= 80:  
        print("Grade: B")  # This will be printed  
    elif score >= 70:  
        print("Grade: C")  
    else:  
        print("Grade: F")  

    4. Nested Conditional Statements:  
    number = 10  
    if number > 0:  
        print("Positive number")  
        if number % 2 == 0:  
            print("Even number")  # This will be printed  
        else:  
            print("Odd number")  
    else:  
        print("Negative number or zero")  

    5. Using Logical Operators: Conditional statements can also use logical operators (and, or, not) to combine multiple conditions.  
    age = 20  
    has_permission = True  
    if age >= 18 and has_permission:  
        print("Access granted.")  # This will be printed  
    else:  
        print("Access denied.")  
    
    6. Using Comparison Operators: Comparison operators (==, !=, <, >, <=, >=) can be used in conditions.  
    num1 = 5  
    num2 = 10  
    if num1 < num2:

7. Describe the different types of loops inb python and their use cases with examples. 

    In Python, loops are used to execute a block of code repeatedly based on a condition. The two primary types of loops are for loops and while loops. Each has its specific use cases and syntax.  

    1. For Loop  
    The for loop is used to iterate over a sequence (like a list, tuple, string, or range). It is commonly used when the number of iterations is known beforehand.  
    Syntax:  
    for variable in sequence:  
         #Code to execute  
        
    Example:  

    Iterating over a list:  
        fruits = ["apple", "banana", "cherry"]  
        for fruit in fruits:  
            print(fruit)  
    
    Output:  
        apple  
        banana  
        cherry  

    Using range():  
        for i in range(5):  
            print(i)  

    Output:  
        0  
        1  
        2  
        3  
        4  
    
    Use Cases:  

    - Iterating through items in a collection (like lists or dictionaries).  
    - Performing operations a specific number of times using range().  

    2. While Loop  
    The while loop continues to execute as long as a specified condition is true. It is useful when the number of iterations is not known in advance and depends on a dynamic condition.  
    Syntax:  
    while condition:  
        #Code to execute  

    Example:  

    Basic while loop:  
        count = 0  
        while count < 5:  
            print(count)  
            count += 1  

    Output:  
        0  
        1  
        2  
        3  
        4  

    Use Cases:  

    - Continuously executing a block of code until a specific condition changes (e.g., reading input until a valid value is provided).  
    - Implementing countdowns or other iterative processes that depend on changing values.  

    3. Nested Loops  
    You can nest loops, meaning you can place one loop inside another. This is often used in more complex scenarios.  

    Example: 

    for i in range(3):  
        for j in range(2):  
            print(f"i: {i}, j: {j}")  

    Output:  
        i: 0, j: 0  
        i: 0, j: 1  
        i: 1, j: 0  
        i: 1, j: 1  
        i: 2, j: 0  
        i: 2, j: 1  

    Use Cases:  
    - Iterating over a matrix or a list of lists.  
    - Performing operations that require multiple levels of iteration.  

    4. Loop Control Statements  
    Additionally, there are control statements that can alter the flow of loops:  

    - break: Exits the loop prematurely. 

    Example:  
    for num in range(10):  
        if num == 5:  
            break  
        print(num)  

    - continue: Skips the current iteration and continues with the next one.

    Example:  
    for num in range(5):  
        if num == 2:  
            continue  
        print(num)   

    - else: Can be used with loops to execute a block of code when the loop completes normally (i.e., not exited with break).

    Example:  
    for num in range(3):  
        print(num)  
    else:  
        print("Loop completed without break.")  

    Summary 

    - for Loop: Best for iterating over a sequence or a known range.  
    - while Loop: Best for situations where the number of iterations is not known in advance.  
    - Nested Loops: Useful for working with multi-dimensional data structures.  
    - Control Statements: Manage loop execution flow with break, continue, and else
